```python
# Visualization of data
# scatter plot sales vs profit
sns.regplot(x='Sales', y='Profit', data=df, scatter_kws={"color": "blue"}, line_kws={"color":"black"})
```

```python
##finding corellation between profit and sales

plt.rcParams['figure.figsize'] = (15,6)

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation matrix for numeric features')
plt.xlabel('sales')
plt.ylabel('profit')

plt.show()
```

```python
## When we look at the correlation coefficient, we see that there is a 48% connection between sales and profit. This is a weak correlation, which means that there are other things that are affecting profit.
#removing unnecessary column such 
df.head()
```

```python
df = df.drop(['Postal Code'], axis=1)
df.head()
```

```python
sns.pairplot(df)
```

```python
#category wise sales and profit
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('PROFIT')
plt.show()
```

```python
##If we analyze the graph above, we can observe that the profit is high for office supplies and technology,but significantly lower for furniture.Therefore, the company should consider these factors andfocus on improving furniture sales.
```

```python
#further divide the category into sub category and then analyze
plt.figure(figsize=(14,10))
statewise = df.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh() # h for horizontal
```

```python
##As we can see most of the furniture items are making loss.
#no of customers- region wise
plt.figure(figsize = (8,8))
plt.title('Region')
plt.pie(df['Region'].value_counts(), labels=df['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
##region wise profit distribution
df.groupby('Region').sum()['Sales']
```

```python
top_states_s = df.groupby("State").Sales.sum().nlargest(n=10)

top_states_p = df.groupby("State").Profit.sum().nlargest(n=10)

plt.style.use('seaborn')
top_states_s.plot(kind = 'bar' ,figsize = (10,6), fontsize = 14)
top_states_p.plot(kind = 'bar' ,figsize = (10,6), fontsize = 14, color = 'red')
plt.xlabel('States')
plt.ylabel('Total Sales')
plt.title('Top 10 States Sales vs Profit', fontsize = 20)
plt.show()
```

```python
# so from the above graph we can analyze that our sells are concentrated in majorily 3 cities , but in the remaining cities there is very less sells. Company needs to focus on the low selling areas.
```
